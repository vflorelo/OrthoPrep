```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.tsv",sep="\t")
```

```python
df = pd.read_csv("data.tsv",sep="\t")
filt_list = ["raw","OP"]
eval_list = ["e-3","e-6"]
mask_list = ["norm","sens"]
inf_list = [1.1,1.5]
fig, axes = plt.subplots(2,4,figsize=(20,10))
counter_1 = 0
counter_2 = 0
for e_value in eval_list:
    for mask in mask_list:
        for inf in inf_list:
            legend_str = "e-value: 1"+e_value+"\nMasking: "+mask+"\nInflation: "+str(inf)
            test_df = df.copy()
            test_df = test_df [(test_df["evalue"]==e_value) & (test_df["mask"]==mask) & (test_df["inf"]==inf)]
            sns.barplot(ax=axes[counter_1,counter_2],data=test_df, x="sp", y="og", hue="filt", gap=0)
            axes[counter_1,counter_2].set_title(legend_str)
            axes[counter_1,counter_2].set_ylim(0, 4500)
            axes[counter_1,counter_2].set(xlabel=None)
            axes[counter_1,counter_2].set(ylabel=None)
            axes[counter_1,counter_2].set(xticklabels=[])
            axes[counter_1,counter_2].set(yticklabels=[])
            counter_2 += 1
            if counter_2%4 == 0:
                counter_2 = 0
                counter_1 += 1
#plt.savefig("orthogroup_distribution.svg")
plt.show()
plt.close()
```
